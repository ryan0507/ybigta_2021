```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
txt = open("Famous-Composers-and-Various.txt", "r").read()
```

```python
txt
```

```python
stopwords
```

```python
bookwc = WordCloud(background_color = "white",
                  collocations = True, 
# 이때 collocations=True 이면 said King, said Hatter 같이 자주 나타나는 단어는 하나의 어구로 분류됨.
                  stopwords = stopwords,
                  colormap = 'flag',
                  min_font_size = 3,
                  width = 600, height = 400)


# alice_novel이라는 텍스트 데이터를 사용하여 wordcloud를 generate하기
bookwc.generate(txt)
bookwc.to_image()
```

```python
bookwc.words_
```

```python
bookwc.to_file("wordcloud.png")
```
